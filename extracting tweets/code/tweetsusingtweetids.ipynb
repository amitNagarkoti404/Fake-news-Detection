```python
#clean the text
#create a function to clean the tweets
def cleantext(text):
  text=re.sub(r'@[A-Za-z0-9]+','',text)     #removing @ mentions
  text=re.sub(r'#','',text)        #removing #
  text=re.sub(r'RT[\s]+','',text)    #removing RT
  text=re.sub(r'https?:\/\/\S+','',text)  # removing hyperlinks
  return text
  
```
